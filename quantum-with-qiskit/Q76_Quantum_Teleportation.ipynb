Asja wants to send a qubit to Balvis by using only classical communication.

Let $ \ket{v} = \myvector{a\\b} \in \mathbb{R}^2 $ be the quantum state.

_Discussion:_ If Asja has many copies of this qubit, then she can collect the statistics based on these qubits and obtain an approximation of $ a $ and $ b $, say $ \tilde{a} $ and $\tilde{b}$, respectively. After this, Asja can send $ \tilde{a} $ and $\tilde{b}$ by using many classical bits, the number of which depends on the precision of the amplitudes. 

On the other hand, If Asja and Balvis share the entangled qubits in state $ \sqrttwo\ket{00} + \sqrttwo\ket{11} $ in advance, then it is possible for Balvis to create $ \ket{v} $ in his qubit after receiving two bits of information from Asja. 

<h3> Protocol </h3>

The protocol uses three qubits as specified below:

<img src='images/quantum_teleportation_qubits.png' width="25%" align="left">

Asja has two qubits and Balvis has one qubit.

Asja's quantum message (key) is $ \ket{v} = \myvector{a\\b} = a\ket{0} + b\ket{1} $.

The entanglement between Asja's second qubit and Balvis' qubit is  $ \sqrttwo\ket{00} + \sqrttwo\ket{11} $.

So, the quantum state of the three qubits is

$$ \mypar{a\ket{0} + b\ket{1}}\mypar{\sqrttwo\ket{00} + \sqrttwo\ket{11}} 
    = \sqrttwo \big( a\ket{000} + a \ket{011} + b\ket{100} + b \ket{111} \big).  $$

<h3> Measurement by Asja </h3>

Asja measures her qubits. With probability $ \frac{1}{4} $, she can observe one of the basis states.

Depeding on the measurement outcomes, Balvis' qubit is in the following states:

- "00": $ \ket{v_{00}} = a\ket{0} + b \ket{1} $ 
- "01": $ \ket{v_{01}} =  a\ket{1} + b \ket{0} $ 
- "10": $ \ket{v_{10}} =  a\ket{0} - b \ket{1} $ 
- "11": $ \ket{v_{11}} =  a\ket{1} - b \ket{0} $ 


As can be observed, the amplitudes $ a $ and $ b $ are "transferred" to Balvis' qubit in each case.

If Asja sends the measurement outcomes, then Balvis can construct $ \ket{v} $ exactly.

```python
#
# your code is here
#

import qiskit.quantum_info as qi
from qiskit.visualization import plot_state_city, array_to_latex, plot_bloch_multivector, plot_state_qsphere
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit, execute, Aer, transpile, execute
import numpy as np
from math import sqrt,pi,sin,cos

simulator = Aer.get_backend('aer_simulator_statevector')

q = QuantumRegister(3)
c2 = ClassicalRegister(1,'c2')
c1 = ClassicalRegister(1,'c1')
circuit = QuantumCircuit(q,c1,c2)

theta = 113

print(theta)

theta = (theta*pi)/180

b = sin(theta)
a = cos(theta)

print("b=",round(b,3),"a=",round(a,3))

circuit.ry(theta,q[2]) #sender code

circuit.h(q[1]) #prepare sender

dm0 = qi.DensityMatrix.from_instruction(circuit)

display(plot_state_city(dm0))

state0 = qi.Statevector(circuit)

print(state0)

display(state0.draw(output="latex"))

display(array_to_latex(state0))

display(plot_bloch_multivector(state0,reverse_bits=True))

circuit.cx(q[1],q[0]) # bind to receiver

dm1 = qi.DensityMatrix.from_instruction(circuit)

display(plot_state_city(dm1))

state1 = qi.Statevector(circuit)

print(state1)

display(array_to_latex(state1))

display(state1.draw(output="latex"))

display(plot_bloch_multivector(state1,reverse_bits=True))

display(plot_state_qsphere(state1.data,show_state_phases=True))

circuit.cx(q[2],q[1]) #sender encode

dm2 = qi.DensityMatrix.from_instruction(circuit)

display(plot_state_city(dm2))

state2 = qi.Statevector(circuit)

print(state2)

display(array_to_latex(state2))

display(state2.draw(output="latex"))

display(plot_bloch_multivector(state2,reverse_bits=True))

display(plot_state_qsphere(state2.data,show_state_phases=True))

circuit.h(q[2])

dm3 = qi.DensityMatrix.from_instruction(circuit)

state3 = qi.Statevector(circuit)

circuit.measure(q[2],c2)
circuit.measure(q[1],c1)

circuit.z(q[0]).c_if(c2,1)
circuit.x(q[0]).c_if(c1,1)

print((c2,c1))

display(circuit.draw("mpl", reverse_bits=True))

display(plot_state_city(dm3))

print(state3)

display(array_to_latex(state3))

display(state3.draw(output="latex"))

display(plot_bloch_multivector(state3,reverse_bits=True))

display(plot_state_qsphere(state1.data,show_state_phases=True))

#job = simulator.run(circuit)
#result = job.result()

job = execute(circuit,Aer.get_backend('statevector_simulator'))
state4 = job.result().get_statevector(circuit)

dm4 = qi.DensityMatrix(state4)

circuit = transpile(circuit, simulator)

display(circuit.draw("mpl", reverse_bits=True))

display(plot_state_city(dm4))

print(state4)

display(array_to_latex(state4))

display(state4.draw(output="latex"))

display(plot_bloch_multivector(state4,reverse_bits=True))

print(job.result().get_counts())
```
